```python
import pandas as pd
import json

input_excel = "synthetic_species_en.xlsx"
output_json = "synthetic_species_en.json"

column_mapping = {
    "id": "id",
    "Sr No": "sr_no",
    "language": "language",
    "Etymology": "etymology",
    "Scientific name": "scientific_name",
    "Common name": "common_name",
    "Habitat": "habitat",
    "Phenology": "phenology",
    "Identification Characters": "identification_characters",
    "Leaf type": "leaf_type",
    "Fruit type": "fruit_type",
    "Seed Germination": "seed_germination",
    "Pest": "pest",
    "Image": "image_url",
    "Video": "videos",
}

schema_order = [
    "id",
    "sr_no",
    "language",
    "etymology",
    "scientific_name",
    "common_name",
    "habitat",
    "phenology",
    "identification_characters",
    "leaf_type",
    "fruit_type",
    "seed_germination",
    "pest",
    "image_url",
    "videos"
]

df = pd.read_excel(input_excel)
df = df.rename(columns=column_mapping)

records = []
for _, row in df.iterrows():
    rec = {}
    for key in schema_order:
        value = row.get(key, None)
        if pd.isna(value):
            value = None
        rec[key] = value
    records.append(rec)

with open(output_json, "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=4)

print("Created the JSON file synthetic data English version")

```
